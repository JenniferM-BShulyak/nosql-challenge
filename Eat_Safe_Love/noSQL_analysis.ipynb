## (2) Which establishments in London have a RatingValue greater than or equal to 4?`

### Find the number of establishments

```python
# Change the type of the 'RatingValue' to Double
establishments.update_many({}, [ {'$set':{'RatingValue': {'$toDouble': '$RatingValue'}}}])
```

```python
# Design query the establishments with LocalAuthority as London and RatingValue greater than or eqaul to 4
query = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': '4'}}
```

```python
# Count the number of establishments that meet the query
print(f"{establishments.count_documents(query)} establishments have a rating of 4 or greater.")
```

```python
pprint(list(establishments.find(query))[0])
```

## (3) What are the top 5 establishments with a RatingValue of '5', sorted by lowest hygiene score nearest to the new restaurant added 'Penang Flavours"?

### Design query

Search within 0.01 degree on either side of the latitude and longitude

Rating value must equal to 5

Sort by hygiene score

```python
# Set the variables
degree = 0.01

# Find the latitude and longitude of 'Penang Flavours' by querying it by name
latitude = establishments.find({'BusinessName': 'Penang Flavours'}, {'geocode.latitude'})[0]['geocode']['latitude']
longitude = establishments.find({'BusinessName': 'Penang Flavours'}, {'geocode.longitude'})[0]['geocode']['longitude']
```

```python
# Set the query
# longitufe and latiude calculations must be done so results are within +- 0.01 degrees
query = {'RatingValue': '5',
        'geocode.latitude': {'$gte': (latitude - degree), '$lte': (latitude + degree)},
        'geocode.longitude': {'$gte': (longitude - degree), '$lte': (longitude + degree)}}

# Sort by the hygiene score
sort = [('scores.Hygiene', -1)]
```

```python
# Run query and print the results
results_3 = list(establishments.find(query).sort(sort))

pprint(results_3[0])
```

### Convert results to dataFrame

```python
df_q3 = pd.DataFrame(results_3)
df_q3.head(10)
```

```python
# How many rows in the dataframe? 
print(f"There are {len(df_q3)} rows in the dataframe")
```

## (4) How many establishments in each Local Authority area have a hygiene score of 0?

### Create pipeline

(1) Matches establishments with a hygiene score of 0

(2) Groups the matches by Local Authority

(3) Sorts the matches from highest to lowest

```python
# query so that the Hygiene score is 0
query_match = {'$match': {'scores.Hygiene': 0}}

# Group by the LocalAuthorityName and sum each additional by 1 with alias of count
query_group = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}

# Sort by highest to lowest
sort_values = {'$sort': {'count': -1}}
```

```python
# set up pipeline
pipeline = [query_match, query_group, sort_values]

```

```python
# run pipeline
results_4 = list(establishments.aggregate(pipeline))
```

```python
# print the num ber of documents in the results
print(f"{len(results_4)} documents are in the results")
```

```python
# Print the first 10 results
for x in range(10):
    pprint(results_4[x])
```

### Convert results to dataFrame

```python
# Convert results to dataframe
df_q4 = pd.DataFrame(results_4)
```

```python
# Display the number of rows in the dataframe
print(f"There are {len(df_q4)} rows in the dataframe")
```
